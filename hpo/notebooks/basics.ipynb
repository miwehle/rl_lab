### Exponential Epsilon Decay

LunarLander:
- 50 steps per second
- ca. 300 Env-Steps pro Episode
- 500 Episoden -> 150_000 steps

```python
# 1
import numpy as np
import plotly.express as px

eps_start = 1.0
eps_end = 0.05
eps_decay = 50_000
steps = np.linspace(0, 150_000, 500)
epsilons = eps_end + (eps_start - eps_end) * np.exp(-steps / eps_decay)

px.line(
    x=steps,
    y=epsilons,
    labels={"x": "Environment steps", "y": "Epsilon"},
)
```

```python
# 2
import numpy as np
import plotly.express as px

eps_start = 1.0
eps_end = 0.05
eps_decay = 50_000
steps = np.linspace(0, 300_000, 500)
epsilons = eps_end + (eps_start - eps_end) * np.exp(-steps / eps_decay)

px.line(
    x=steps,
    y=epsilons,
    labels={"x": "Environment steps", "y": "Epsilon"},
)
```

```python
# 3
import plotly.graph_objects as go

figure = go.Figure()
for episodes, max_steps, dash in [
    (1_000, 300_000, "solid"),
    (500, 150_000, "dash"),
]:
    steps = np.linspace(0, max_steps, 500)
    epsilons = eps_end + (eps_start - eps_end) * np.exp(-steps / eps_decay)
    figure.add_scatter(
        x=steps,
        y=epsilons,
        name=f"{episodes} episodes",
        line={"dash": dash},
    )

figure.update_layout(
    xaxis_title="Environment steps",
    yaxis_title="Epsilon",
)
figure.show()
```
